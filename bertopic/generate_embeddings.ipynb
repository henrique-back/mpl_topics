```python
import numpy as np
import pandas as pd

# Set embedding model
from sentence_transformers import SentenceTransformer
sentence_model = SentenceTransformer("allenai-specter")

# Prepare the data
data = pd.read_csv(r'..\alldata.csv')

#Create abstract dataframe (docs) and save (change path to suit your case)
docs = data['Abstract']
docs = docs.tolist()

# Prepare embeddings and save (this saves a lot of time in the future)
embeddings = sentence_model.encode(docs, show_progress_bar=True)
np.save(r'..\bertopic\preprocessed_data\embeddings.csv.npy', embeddings)
print(embeddings.shape)
```

```python
# Visualize a heatmap of the embeddings. This is done to get a sense of the dimensions and values of the embedded data

import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(20, 16))

# Plot the heatmap
plt.imshow(embeddings)

# Adjust the aspect ratio of the plot
plt.gca().set_aspect('auto')

# Add a colorbar to the right
plt.colorbar()

# Show the plot
plt.show()
```
